```python
import pandas as pd

jt_year_theme = pd.read_csv("../data/processed/jt_topics_by_year_theme.csv")
jt_year_channel_theme = pd.read_csv("../data/processed/jt_topics_by_year_channel_theme.csv")
```

```python
jt_year_theme.head()
jt_year_channel_theme.head()
```

```python
top_themes_global = (
    jt_year_theme.groupby("theme", as_index=False)
    .agg(
        total_subjects_all_years=("total_subjects", "sum"),
        total_duration_all_years=("total_duration", "sum"),
    )
    .sort_values("total_subjects_all_years", ascending=False)
)

top_themes_global.head(10)
```

```python
top_themes_global.sort_values("total_duration_all_years", ascending=False).head(10)
```

```python
top_themes_by_channel = (
    jt_year_channel_theme.groupby(["channel_name", "theme"], as_index=False)
    .agg(
        total_subjects_all_years=("total_subjects", "sum"),
        total_duration_all_years=("total_duration", "sum"),
    )
)

top5_by_channel = (
    top_themes_by_channel.sort_values(
        ["channel_name", "total_subjects_all_years"],
        ascending=[True, False]
    )
    .groupby("channel_name")
    .head(5)
)

top5_by_channel
```
