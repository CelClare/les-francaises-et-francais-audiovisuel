Vision globale des 4 datasets choisis sur les 11 datasets du sujet.

```python
import pandas as pd
from pathlib import Path

files = list(Path("../data/raw").glob("*"))

for f in files:
    print("="*80)
    print("Nom :", f.name)
    
    if f.suffix == ".csv":
        try:
            df = pd.read_csv(f, nrows=5)  # tentative en UTF-8 par défaut
        except UnicodeDecodeError:
            df = pd.read_csv(f, sep=';', encoding="latin1", nrows=5)
        
    elif f.suffix == ".xlsx":
        try:
            df = pd.read_excel(f, nrows=5)
        except Exception as e:
            print(f"Erreur lors de la lecture du fichier {f.name} : {e}")
    else:
        continue
        
    print("Colonnes :", df.columns.tolist())
    print("Nb colonnes :", len(df.columns))
    print()
```

Création d'indicateurs potentiels sur la part féminine dans les programmes, la part masculine, etc.

```python
# Créer une durée totale genrée
df_tv["total_gender_duration"] = df_tv["male_duration"] + df_tv["female_duration"]
# Créer une part féminine
df_tv["female_share"] = df_tv["female_duration"] / df_tv["total_gender_duration"]
# Créer une part masculine
df_tv["male_share"] = df_tv["male_duration"] / df_tv["total_gender_duration"]
# Vérifier les nouvelles colonnes
df_tv[["male_duration", "female_duration", "total_gender_duration", "female_share", "male_share"]].head()
```

Une première comparaison entre chaînes publiques et privées montre que la part féminine moyenne est plus élevée sur les chaînes publiques (33,1 %) que sur les chaînes privées (28,3 %) sur le périmètre étudié.
Cette différence suggère une sous-représentation plus marquée des durées féminines sur les chaînes privées, selon l’indicateur construit.
Une analyse plus fine par chaîne montre toutefois des écarts importants à l’intérieur de chaque groupe.

```python
df_tv.groupby("is_public_channel")[["female_share"]].agg(["mean", "median"])
```
